```python
def categoria(edad):
    if 18 < edad <= 25:
        return 'Joven'
    elif 26 < edad <=35:
        return 'Adulto joven'
    elif 26 < edad <= 45:
        return 'Adulto'
    elif 46 < edad <=55:
        return 'Senior'
    else:
        return 'No califica'
```
