```python
# Importamos las librerías que necesitamos
import numpy as np
# Librerías de extracción de datos
# -----------------------------------------------------------------------
from bs4 import BeautifulSoup
import requests

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd
```

```python
### Paso 1
url_ebay = 'https://www.ebay.es/e/_moda/adidassneakerses?_pgn=1'
### Paso 2
res_zapatillas = requests.get(url_ebay)
### Paso 3
print(res_zapatillas.status_code)

# Paso 4
sopa_zapatillas = BeautifulSoup(res_zapatillas.content, 'html.parser')  ### traducir del html
```

```python
#sopa_zapatillas ###ahora usamos la inspección
### buscamos lo que nos interesa
```

```python
### tomamos el código de la inspección
# Paso 5
lista_zapas = sopa_zapatillas.findAll('h3', {'class': 's-item__title'} )
lista_zapas[:2]
```

```python
nombre_zapatillas = [zapatilla.getText() for zapatilla in lista_zapas]
```

```python
lista_zapas[0].getText()
```

```python
nombre_zapatillas
```

```python
lista_precios_zapas = sopa_zapatillas.findAll('span', {'class': 's-item__price'} )
lista_precios_zapas
```

```python
precios_zapatillas = [precio.getText() for precio in lista_precios_zapas]
```

```python
precios_zapatillas[0]
```

```python
###np.where es como un if...else 

precios_zapatillas_df = pd.Series(precios_zapatillas)
precios_zapatillas_df = precios_zapatillas_df.str.split(" a ",expand=True)
precios_zapatillas_df.columns = ["precio_min","precio_max"]
precios_zapatillas_df["precio_max"] = np.where(precios_zapatillas_df["precio_max"].isna(),precios_zapatillas_df["precio_min"], precios_zapatillas_df["precio_max"])
precios_zapatillas_df = precios_zapatillas_df.applymap(lambda x: x.replace("EUR","").replace(",",".")).astype(float)
precios_zapatillas_df.head()
```

```python
df_precios = pd.DataFrame(precios_zapatillas, columns=['Precio'])
df_precios.head()
```

```python
df_precios = df_precios['Precio'].str.split('a', expand=True) ### expand divide en dos 2
df_precios.head()
```

```python
df_zapatillas = precios_zapatillas_df.copy()
df_zapatillas['nombre'] = nombre_zapatillas ### creo columna de la lista
```

```python
df_zapatillas.head()
```

```python
df_zapatillas = df_zapatillas.reindex(columns=['nombre', 'precio_min', 'precio_max'])
df_zapatillas.head(1)
```

```python
df_zapatillas.insert(0,'nombre_zap', value= nombre_zapatillas) ### de esta forma creo una columna en el lugar que quiero y con los valores de lista que quiero
```

```python
df_zapatillas['diferencia'] = df_zapatillas['precio_max'] - df_zapatillas['precio_min']
df_zapatillas.head(1)
```

```python
lista_precio_envio = sopa_zapatillas.findAll('span', {'class': 's-item__shipping s-item__logisticsCost'})
lista_precio_envio.head(1)
```

```python
from datetime import datetime
df_zapatillas['fecha_recoleccion'] = datetime.now()
```

```python
df_zapatillas.head(1)
```

```python
df_zapatillas.to_csv(f'datos/datos_ebay_zapatillas{datetime.now().date()}.csv')
### se guarda con todo lo de ahora 
```
